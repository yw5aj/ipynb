```python
from sympy import symbols, sqrt, simplify, Matrix, Q, ask, refine, S
from sympy.assumptions.assume import global_assumptions

v11, v22, v33, v12, v13, v23 = symbols('v11, v22, v33, v12, v13, v23')
v = Matrix([[v11, v12, v13], [v12, v22, v23], [v13, v23, v33]])
for elem in v:
    global_assumptions.add(Q.real(elem))
global_assumptions.add(Q.positive_definite(v))
```

```python
J = v.det()
global_assumptions.add(Q.positive(J))
lambda1, lambda2, lambda3 = v.eigenvals()
global_assumptions.add(Q.positive(lambda1))
global_assumptions.add(Q.positive(lambda2))
global_assumptions.add(Q.positive(lambda3))
```

```python
lambdabar1 = J**(-S(1)/3) * lambda1
lambdabar2 = J**(-S(1)/3) * lambda2
lambdabar3 = J**(-S(1)/3) * lambda3
```
