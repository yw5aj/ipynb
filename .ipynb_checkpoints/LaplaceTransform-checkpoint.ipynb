```python
from sympy import symbols, exp
from sympy.integrals import laplace_transform, inverse_laplace_transform

t, s = symbols('t, s')
E1, E2, eta1 = symbols('E1, E2, eta1', positive=True)

q0 = E1 * E2 / (E1 + E2)
p1 = eta1 / (E1 + E2)
q1 = E2 * eta1 / (E1 + E2)
```

```python
tau1 = eta1 / E1
Jt = 1 / E2 + 1 / E1 * (1 - exp(-t / tau1))
Js = laplace_transform(Jt, t, s)[0]
```

```python
inverse_laplace_transform(1 / Js / s**2, s, t)
```

```python
Yt = q0 + (q1 / p1 - q0) * exp (-t / p1)
Ys = laplace_transform(Yt, t, s)[0]
```
