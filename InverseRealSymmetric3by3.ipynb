```python
from sympy import symbols, Matrix, fcode
C11, C22, C33, C12, C13, C23 = symbols('C11, C22, C33, C12, C13, C23', real=True)
```

```python
C = Matrix(((C11, C12, C13), (C12, C22, C23), (C13, C23, C33)))
```

```python
C.inv()
```

```python
F11, F12, F13, F21, F22, F23, F31, F32, F33 = symbols('F11, F12, F13, F21, F22, F23, F31, F32, F33', real=True)
F = Matrix([[F11, F12, F13], [F21, F22, F23], [F31, F32, F33]])
Finv = F.inv()
```

```python
import numpy as np
Finv = np.array(Finv)
Finv
```

```python
# Replace into code format
result = str(Finv.ravel('F'))
replist = [('F11', 'f(1, 1)'), ('F12', 'f(1, 2)'), ('F13', 'f(1, 3)'),
           ('F21', 'f(2, 1)'), ('F22', 'f(2, 2)'), ('F23', 'f(2, 3)'),
           ('F31', 'f(3, 1)'), ('F32', 'f(3, 2)'), ('F33', 'f(3, 3)'),
           ('\n', '')]
for old, new in replist:
    result = result.replace(old, new)
```

```python
# Break into lines
def breaklines(old, length=70):
    new = '&\n'.join(old[i:i + length] for i in range(0, len(old), length))
    return new
result = breaklines(result)
```
